```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the list of diseases to scrape
diseases = ['Cholera', 'Malaria', 'Dengue', 'Zika', 'Yellow Fever', 'HIV/AIDS', 'Tuberculosis', 'Measles', 'Rubella', 'Polio']

# Create a DataFrame to store the scraped data
df = pd.DataFrame()

# Loop through the diseases list
for disease in diseases:

    # Get the Wikipedia URL for the disease
    url = 'https://en.wikipedia.org/wiki/' + disease

    # Make a request to the URL
    response = requests.get(url)

    # Parse the response into a BeautifulSoup object
    soup = BeautifulSoup(response.content, 'html.parser')

    try:
        # Find the table containing the disease information
        table = soup.find('table', class_='infobox vcard')
    except AttributeError:
        pass

    # Extract the disease name
    if table is not None:
        name = table.find('th', class_='fn').text

    # Extract the disease description
    if table is not None:
        description = table.find('td', class_='description').text

    # Extract the disease symptoms
    if table is not None:
        symptoms = table.find('td', class_='symptoms').text

    # Extract the disease causes
    if table is not None:
        causes = table.find('td', class_='causes').text

    # Extract the disease diagnosis
    if table is not None:
        diagnosis = table.find('td', class_='diagnosis').text

    # Extract the disease treatment
    if table is not None:
        treatment = table.find('td', class_='treatment').text

    # Add the scraped data to the DataFrame
    if table is not None:
        df = df.append({'Disease': name, 'Description': description, 'Symptoms': symptoms, 'Causes': causes, 'Diagnosis': diagnosis, 'Treatment': treatment}, ignore_index=True)

# Print the DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create a list of diseases.
diseases = ["Cholera", "Malaria", "Dengue", "Yellow fever", "Zika", "HIV/AIDS", "Tuberculosis", "Measles", "Rubella", "Polio"]

# Create a function to scrape the information for each disease.
def scrape_disease(disease):
    url = "https://en.wikipedia.org/wiki/" + disease
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Get the disease name.
    disease_name = soup.find("h1").text

    # Get the disease description.
    disease_description = soup.find("div", class_="mw-content-ltr").text

    # Get the disease symptoms.
    disease_symptoms = soup.find("ul", class_="mw-list-of-links").find_all("li")

    # Get the disease causes.
    disease_causes = soup.find("ul", class_="mw-list-of-links").find_all("li")

    # Get the disease treatments.
    disease_treatments = soup.find("ul", class_="mw-list-of-links").find_all("li")

    # Create a dictionary to store the information.
    disease_info = {
        "name": disease_name,
        "description": disease_description,
        "symptoms": disease_symptoms,
        "causes": disease_causes,
        "treatments": disease_treatments
    }

    return disease_info

# Use a for loop to iterate through the list of diseases and scrape the information for each disease.
disease_infos = []
for disease in diseases:
    disease_info = scrape_disease(disease)
    disease_infos.append(disease_info)

# Save the information in a DataFrame.
df = pd.DataFrame(disease_infos)
df.to_csv("diseases.csv", index=False)
```

```python
import re

def clean_text(text):
    # Remove HTML tags
    cleaned_text = re.sub('<.*?>', '', text)
    # Remove punctuation and special characters
    cleaned_text = re.sub('[^\w\s]', '', cleaned_text)
    # Remove non-alphanumeric characters
    cleaned_text = re.sub('\W+', ' ', cleaned_text)
    # Remove extra whitespaces
    cleaned_text = re.sub('\s+', ' ', cleaned_text)
    
    return cleaned_text
```
