```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create a list of diseases.
diseases = ["Cholera", "Malaria", "Dengue", "Yellow fever", "Zika", "HIV/AIDS", "Tuberculosis", "Measles", "Rubella", "Polio"]

# Create a function to scrape the information for each disease.
def scrape_disease(disease):
    url = "https://en.wikipedia.org/wiki/" + disease
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Get the disease name.
    disease_name = soup.find("h1").text

    # Get the disease description.
    disease_description = soup.find("div", class_="mw-content-ltr").text

    # Get the disease symptoms.
    disease_symptoms = soup.find("ul", class_="mw-list-of-links").find_all("li")

    # Get the disease causes.
    disease_causes = soup.find("ul", class_="mw-list-of-links").find_all("li")

    # Get the disease treatments.
    disease_treatments = soup.find("ul", class_="mw-list-of-links").find_all("li")

    # Create a dictionary to store the information.
    disease_info = {
        "name": disease_name,
        "description": disease_description,
        "symptoms": disease_symptoms,
        "causes": disease_causes,
        "treatments": disease_treatments
    }

    return disease_info

# Use a for loop to iterate through the list of diseases and scrape the information for each disease.
disease_infos = []
for disease in diseases:
    disease_info = scrape_disease(disease)
    disease_infos.append(disease_info)

# Save the information in a DataFrame.
df = pd.DataFrame(disease_infos)
df.to_csv("diseases.csv", index=False)
```

```python
import requests
from bs4 import BeautifulSoup

# Disease link to scrape
disease_link = "https://www.mayoclinic.org/diseases-conditions/fever/symptoms-causes/syc-20352759"

# Function to scrape data from the disease link
def scrape_data_from_link(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the main content section where the data is located
    main_content = soup.find("div", class_="content")

    # Find the <h2> element
    h2_element = main_content.find("h2")

    # Find the <div> element under the <h2> element
    div_element = h2_element.find_next_sibling("div")

    # Find the <ul> element after the <div> element
    ul_element = div_element.find_next_sibling("ul")

    # Find all <li> elements under the <ul> element
    li_elements = ul_element.find_all("li")

    # Extract the text from the <p> element after the <ul> element
    p_element = ul_element.find_next_sibling("p")
    p_text = p_element.get_text()

    # Concatenate the data into a single string
    data = f"{h2_element.get_text()}\n\n{div_element.get_text()}\n\n"
    data += "\n".join([li.get_text() for li in li_elements])
    data += f"\n\n{p_text}"

    return data

# Scrape the data for the Fever disease
fever_data = scrape_data_from_link(disease_link)

# Print or do further processing with the scraped data
print(fever_data,li_elements)
```

```python
import requests
from bs4 import BeautifulSoup

# Disease link to scrape
disease_link = "https://www.mayoclinic.org/diseases-conditions/fever/symptoms-causes/syc-20352759"

# Function to scrape text from the disease link
def scrape_text_from_link(link):
    response = requests.get(link)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the main content section where the text is located
    main_content = soup.find("div", class_="content")

    # Find all <h2> elements and extract the text
    h2_elements = main_content.find_all("h2")
    h2_text = [h2.get_text() for h2 in h2_elements]

    # Find all <p> elements and extract the text
    p_elements = main_content.find_all("p")
    p_text = [p.get_text() for p in p_elements]

    # Find all <li> elements and extract the text
    li_elements = main_content.find_all("li")
    li_text = [li.get_text() for li in li_elements]

    # Concatenate the extracted text into a single string
    text = "\n".join(h2_text + p_text + li_text)

    return text

# Scrape the text information for the Fever disease
fever_text = scrape_text_from_link(disease_link)

# Print or do further processing with the scraped text
print(fever_text)
```

```python
import re

def clean_text(text):
    # Remove HTML tags
    cleaned_text = re.sub('<.*?>', '', text)
    # Remove punctuation and special characters
    cleaned_text = re.sub('[^\w\s]', '', cleaned_text)
    # Remove non-alphanumeric characters
    cleaned_text = re.sub('\W+', ' ', cleaned_text)
    # Remove extra whitespaces
    cleaned_text = re.sub('\s+', ' ', cleaned_text)
    
    return cleaned_text
```

```python
from nltk.corpus import stopwords

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    return filtered_tokens
```
